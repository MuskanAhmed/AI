```python
from random import *
import random
li1 = []
n=int(input("enter player numbers: "))
totaltarget=int(input("enter total target: "))
for i in range(n):
  alist=[]
  playername=input("enter player name: ")
  alist.append(playername)
  playerrun=input("enter run: ")
  alist.append(playerrun)
  li1.append(alist)

# print(li1)

#EXAMPLE:


# n=8                                                              
# totaltarget=330
# li1=[['Tamim', '68'], ['Shoumyo', '25'], ['Shakib', '70'], ['Afif', '53'], ['Mushfiq', '71'], ['Liton', '55'], ['Mahmudullah', '66'], ['Shanto', '29']]


# n=5
# totaltarget=240
# li1=[['Bradman', '120'], ['Tendulkar', '90'], ['Sangakkara', '70'], ['Kallis', '65'], ['Lara', '80']]


#n=4
#totaltarget=100
#li1=[['Ralph', '80'], ['John', '70'], ['Tom', '40'], ['Young', '50']]



def population(l):
  to_population=[]
  for i in range(30):
    randlis = lambda n: [randint(0,1) for b in range(1,n+1)]
    temp=randlis(l)
    to_population.append(temp)
  return to_population

def fitnesscalc(mhold,li1):
  fitnessvalues=[]
  for i in mhold:
    temp=0
    for j in range(len(i)):
      if i[j]==1:
        temp+=int(li1[j][1])
    fitnessvalues.append(temp)
  return fitnessvalues

def selection(fit,mhold):
  fitn=[]
  mtemp=[]
  mini=min(fit)
  idx=fit.index(mini)
  fit.pop(idx)
  mhold.pop(idx)
  fitn=fit
  mtemp=mhold
  return fitn, mtemp

def crossing(x,y):
  temp=random.choice(range(1,len(x)-1))
  x1=x[0:temp:]
  x2=x[temp::]
  y1=y[0:temp:]
  y2=y[temp::]
  aftercrossing_x=x1+y2
  aftercrossing_y=y1+x2
  return aftercrossing_x, aftercrossing_y

def mutation(child):
  temp=random.choice(range(len(child)))
  if child[temp]==0:
    child[temp]=1
  else:
    child[temp]=0
  return child

target= False
mhold=population(n)
#print(mhold)
fit=fitnesscalc(mhold,li1)
#print(fit)
m1,m2=selection(fit,mhold)
for i in m1:
  if i==totaltarget:
    indx=m1.index(i)
    mtem=[]
    for i in range(len(li1)):
      mtem.append(li1[i][0])
    print(mtem)
    print(m2[indx])
    target= True
    

if target==False:
  for i in range(200):
    newlis=[]
    for i in range(len(m2)):
      x = random.choice(m2)
      y = random.choice(m2)
      child1,child2 = crossing(x, y)
      mutatedchild1=mutation(child1)
      newlis.append(mutatedchild1)
      mutatedchild2=mutation(child2)
      newlis.append(mutatedchild2)
    #print(len(newlis))
    
    newfitness=fitnesscalc(newlis,li1)
    ma1,ma2=selection(newfitness,newlis)
    #print(ma1)
    for i in ma1:
      if i==totaltarget:
        indx=ma1.index(i)
        target= True
        #print(indx)
    if target==True:
      break
  if target==True:
    mtem=[]
    for i in range(len(li1)):
      mtem.append(li1[i][0])
    print(mtem)
    print(ma2[indx])
  else:
    print(-1)
```
