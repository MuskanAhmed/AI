```python
import math
import random
user_input=input("Enter your student ID: ")
new_id=""
for i in user_input:
  if i!="0":
    new_id+=i
  else:
    new_id+="8"
#print(new_id)
minimum=int(new_id[4])
total_points_to_win=int(new_id[-1:-3:-1])
maximum=math.ceil(total_points_to_win*1.5)
total_number_of_shuffle=int(new_id[3])
mhold=[random.randint(minimum,maximum) for i in range(8)]
print("Generated 8 random points between the minimum and maximum point limits:",mhold)
print("Total points to win:",total_points_to_win)
#print(total_number_of_shuffle)
arr=[]
def alpha_beta(depth,node,maxim,alist,alp,beta):
  if depth==3:
    return alist[node]
  if maxim:
    temp=minimum
    for i in range(2):
      val=alpha_beta(depth+1,node*2+i,False,alist,alp,beta)
      temp=max(temp,val)
      alp=max(alp,temp)
      if beta<=alp:
        break
    return temp
  else:
    temp=maximum
    for i in range(2):
      val=alpha_beta(depth+1,node*2+i,True,alist,alp,beta)
      temp=min(temp,val)
      beta=min(beta,temp)
      if beta<=alp:
        break
    return temp
def shuffle(mhold,arr,total_number_of_shuffle):
  for i in range(total_number_of_shuffle):
    random.shuffle(mhold)
    temp=alpha_beta(0,0,True,mhold,minimum,maximum)
    arr.append(temp)
  return arr
point=(alpha_beta(0,0,True,mhold,minimum,maximum))
print("Achieved point by applying alpha-beta pruning =",point)
if point>=total_points_to_win:
  print("The winner is Optimus Prime")
else:
  print("The winner is Megatron")
shuffledlist=shuffle(mhold,arr,total_number_of_shuffle)
print("\nAfter the shuffle:")
print("List of all points values from each shuffle:",shuffledlist)
mtemp=[]
for j in shuffledlist:
  if j>=total_points_to_win:
    mtemp.append(j)
amax=max(shuffledlist)
mnum=len(mtemp)
print("The maximum value of all shuffles:",amax)
print("Won",mnum,"times out of",total_number_of_shuffle, "number of shuffles")
```
